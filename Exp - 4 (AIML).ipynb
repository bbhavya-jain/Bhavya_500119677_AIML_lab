```python
import pandas as pd

# Function to load dataset (import)
def load_dataset(file_path):
    # Load the dataset from a CSV file
    data = pd.read_csv('rental-september.csv')
    return data

# Function to export dataset
def export_dataset(data, file_path):
    # Save the dataframe to a CSV file
    data.to_csv(file_path, index=False)
    print(f"Dataset exported successfully to {file_path}")

# Function to show dataset details
def show_dataset_details(data):
    # Number of rows and columns
    print(f"Number of rows: {data.shape[0]}")
    print(f"Number of columns: {data.shape[1]}")

    # First five rows of the dataset
    print("\nFirst five rows:")
    print(data.head())

    # Dataset size (number of elements)
    print(f"\nDataset size (number of elements): {data.size}")

    # Count of missing values in each column
    print("\nNumber of missing values in each column:")
    print(data.isnull().sum())

    # Summary of numerical columns
    print("\nNumerical columns summary:")
    print("Sum:\n", data.sum(numeric_only=True))
    print("Average:\n", data.mean(numeric_only=True))
    print("Minimum values:\n", data.min(numeric_only=True))
    print("Maximum values:\n", data.max(numeric_only=True))

# Example usage
if __name__ == "__main__":
    # Import data (Assuming a CSV file path)
    file_path = 'your_dataset.csv'
    data = load_dataset(file_path)

    # Show details of the dataset
    show_dataset_details(data)

    # Export the dataset (optional)
    export_file_path = 'exported_dataset.csv'
    export_dataset(data, export_file_path)
```
